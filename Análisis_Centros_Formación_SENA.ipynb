```python
import pandas as pd
import numpy as np
```

```python
# URL del archivo CSV
url = 'https://raw.githubusercontent.com/devyizuz/datasetsena/main/Georeferenciaci_n_Centros_de_Aprendizaje_SENA_20240515.csv'

# Leemos los datos desde la URL con pandas
df = pd.read_csv(url)
df.head()
```

```python
# Verificamos los tipos de datos
print(df.dtypes)
```

```python
# Contamos el número de registros (filas) en el DataFrame
num_registros = df.shape[0]
print(f"El número de registros en el dataset es: {num_registros}")
```

```python
# Verificamos si hay valores nulos
print(df.isnull().sum())
```

```python
# Verificamos si hay valores duplicados en una columna específica, por ejemplo, 'CENTRO_FORMACION'
duplicados_centro_formacion = df[df.duplicated(subset=['CODIGO_CENTRO'])]
print("Filas con 'CENTRO_FORMACION' duplicado:")
print(duplicados_centro_formacion)
```

```python
# Contamos si hay valores duplicados en el DataFrame completo
print(df.duplicated().sum())
```

```python
# Verificamos cuales valores estan duplicados en el DataFrame completo
duplicados = df[df.duplicated()]
print("Filas duplicadas:")
print(duplicados)
```

```python
# Eliminamos los duplicados en todo el DataFrame y mantenemos la primera ocurrencia
df_sin_duplicados = df.drop_duplicates()

print("DataFrame sin duplicados:")
print(df_sin_duplicados)
```

```python
# Contamos el número de registros (filas) en el DataFrame
num_registros = df_sin_duplicados.shape[0]
print(f"El número de registros en el dataset es: {num_registros}")
```

```python
# Estandarizamos los nombres de los centros a mayúsculas
df_sin_duplicados.loc[:, 'CENTRO_FORMACION'] = df_sin_duplicados['CENTRO_FORMACION'].str.upper()

# Unificamos los formatos de los textos eliminando espacios innecesarios
df_sin_duplicados.loc[:, 'NOMBRE_MUNICIPIO'] = df_sin_duplicados['NOMBRE_MUNICIPIO'].str.strip()
df_sin_duplicados.loc[:, 'NOMBRE_DEPARTAMENTO'] = df_sin_duplicados['NOMBRE_DEPARTAMENTO'].str.strip()

df_sin_duplicados.head()
```

```python
# Verificamos que no queden valores nulos
print(df_sin_duplicados.isnull().sum())

# Verificamos que no queden duplicados
print(df_sin_duplicados.duplicated().sum())
```

```python
# Número de centros por departamento
# Agrupamos por 'NOMBRE_DEPARTAMENTO' y contamos los centros de formación
numero_centros_por_departamento = df_sin_duplicados.groupby('NOMBRE_DEPARTAMENTO')['CENTRO_FORMACION'].count()

# Ordenamos de mayor a menor
numero_centros_ordenados = numero_centros_por_departamento.sort_values(ascending=False)

# Imprimimos los resultados ordenados
print(numero_centros_ordenados)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Verificamos que df_sin_duplicados y la columna NOMBRE_REGIONAL existan
if 'df_sin_duplicados' in locals() and 'NOMBRE_REGIONAL' in df_sin_duplicados.columns:
    # Ordenar por la cuenta de cada región de mayor a menor
    region_counts = df_sin_duplicados['NOMBRE_REGIONAL'].value_counts().sort_values(ascending=False)
    sns.countplot(data=df_sin_duplicados, x='NOMBRE_REGIONAL', order=region_counts.index)
    plt.xticks(rotation=90)
    plt.show()
else:
    print("Verifique que 'df_sin_duplicados' y la columna 'NOMBRE_REGIONAL' existan.")

# Verificar que 'df_sin_duplicados' esté definido y que las columnas 'LATITUD' y 'LONGITUD' existan en 'df_sin_duplicados'
if 'df_sin_duplicados' in locals() and isinstance(df_sin_duplicados, pd.DataFrame) and 'LATITUD' in df_sin_duplicados.columns and 'LONGITUD' in df_sin_duplicados.columns:
    # Análisis de correlación entre latitud y longitud
    correlacion = df_sin_duplicados[['LATITUD', 'LONGITUD']].corr()
    print(correlacion)
else:
    print("Verifique que 'df_sin_duplicados' esté definido como DataFrame y que las columnas 'LATITUD' y 'LONGITUD' existan.")
```
